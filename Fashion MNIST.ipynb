```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
```

```python
layer1 = Dense(units =1, input_shape = [1])

model = Sequential(layer1)

model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
x = np.array([-1,0,1,2,3,4])
y = np.array([-3,-1,1,3,5,7])
model.fit(x,y,epochs = 500)
```

```python
model.predict([10])
```

```python
from tensorflow.keras.layers import Flatten
```

```python
from tensorflow.python.ops.gen_nn_ops import softmax
model = Sequential([
                    Flatten(input_shape = (28,28)),
                    Dense(units = 128, activation = tf.nn.relu),
                    Dense(units = 10, activation = tf.nn.softmax)
])
```

```python
data = tf.keras.datasets.fashion_mnist
```

```python
(tr_images,tr_lables), (t_images,t_labels) = data.load_data()
```

```python
tr_images = tr_images/255.0
t_images = t_images/255.0
#NORMALISING THE IMAGE ARRAY
```

```python
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(tr_images,tr_lables, epochs = 5)
```

```python
 model.evaluate(t_images,t_labels)
```

```python
p_labels = model.predict(t_images)
```

```python
print(p_labels[0])  #gives back the output at neuron after softmax
```

```python
print(t_labels[0])  # test label, can be seen that 9th index of prediction has highest probability
```

```python
# Using Call backs inorder to get more deep accuracy without overfitting
```

```python
class mycallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self,epoch,logs={}):
    if(logs.get('accuracy')>0.95):
      print('Maximum attained accuracy acheived, Stopping the training')
      self.model.stop_training = True

callback = mycallback()
model.fit(tr_images,tr_lables,epochs = 50,callbacks = [callback])
```

```python
model.evaluate(t_images,t_labels)
```

```python
# Using Convolutions to learn features of the images
```

```python
tr_images = tr_images.reshape(60000,28,28,1)
t_images = t_images.reshape(10000,28,28,1)
```

```python
modelConv = Sequential([
                        tf.keras.layers.Conv2D(64,(3,3),activation = 'relu', input_shape=(28,28,1)),
                        tf.keras.layers.MaxPooling2D(2,2),
                        tf.keras.layers.Conv2D(64,(3,3),activation = 'relu'),
                        tf.keras.layers.MaxPooling2D(2,2),
                        tf.keras.layers.Flatten(),
                        tf.keras.layers.Dense(128,activation = tf.nn.relu),
                        tf.keras.layers.Dense(10,activation = tf.nn.softmax)

])
```

```python
modelConv.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
```

```python
modelConv.fit(tr_images,tr_lables,epochs= 10)
```

```python
classificationConv2d = modelConv.predict(t_images)
```

```python
print(classificationConv2d[0])
```

```python
print(t_labels[0])
```

```python
# Can be clearly seen that 9th Index has the probability of test input which is correct as per the test label.
```
